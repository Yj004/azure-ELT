```
# Define the storage account and access key
blob_account_name = "tokyoazureyashjeet"
blob_container_name = "tokyo-olmypic"  # Corrected container name
blob_file_name = "startup_funding.csv"

# Set up the configuration to access Blob Storage
spark.conf.set(
    f"fs.azure.account.key.{blob_account_name}.dfs.core.windows.net",
    "secret-key"
)

# Load data from Azure Blob Storage
df = spark.read.csv(
    f"abfss://{blob_container_name}@{blob_account_name}.dfs.core.windows.net/{blob_file_name}",
    header=True,
    inferSchema=True
)
display(df)
```

```
from pyspark.sql.functions import col, sum as _sum

null_counts = df.select([
    _sum(col(c).isNull().cast("int")).alias(c) for c in df.columns
])
display(null_counts)

df_cleaned = df.dropna()

# Display the cleaned dataframe
display(df_cleaned)
```

```
# Define the output path
output_path = f"abfss://{blob_container_name}@{blob_account_name}.dfs.core.windows.net/cleaned_startup_funding.csv"

# Save the cleaned DataFrame to Azure Blob Storage
df_cleaned.write.csv(output_path, header=True, mode="overwrite")
```

```
# Create a temporary view of the DataFrame
df.createOrReplaceTempView("startup_funding")

# SQL query to filter for the fintech industry vertical
fintech_df = spark.sql("""
SELECT `Investors Name`
FROM startup_funding
WHERE `Industry Vertical` = 'Fintech'
""")

# Display the filtered DataFrame
display(fintech_df)
```
